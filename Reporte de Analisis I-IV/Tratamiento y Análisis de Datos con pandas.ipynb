**seleccion_casas:** es una variable que almacena una condición **booleana**. Esta condición se evalúa como **True** para cada fila en el DataFrame df_residencias_2 si el valor en la columna "Tipo" es igual a "Casa", "Casa en condominio" o "Casa de villa". Es decir, seleccion_casas será True para las filas que representan casas en cualquiera de esas tres categorías y False para las demás filas.

**casas:** es una nueva variable que almacena un DataFrame que contiene únicamente las filas de df_residencias_2 que cumplen con la condición especificada en seleccion_casas. En resumen, casas es un subconjunto de df_residencias_2 que contiene solo las filas que representan casas en alguna categoria de las ya mencionadas.

En este código, se filtran y se seleccionan las casas en base a sus categorías de tipo y se almacenan en la variable casas.






```python
seleccion_casas = (df_residencias_2["Tipo"] == "Casa") | \
 (df_residencias_2["Tipo"] == "Casa en condominio") | \
  (df_residencias_2["Tipo"] == "Casa de villa")
casas = df_residencias_2[seleccion_casas] #"|" => OR
casas
```

```python
casas = df_residencias_2[seleccion_casas].shape[0]
casas
```

De esta forma me trae todos los registros que en la columna valor figuran valores nulos.

```python
dataset_alquileres[dataset_alquileres["Valor"].isnull()]
```

Eliminando todos los registros que contengan valores nulos en la columna valor.

**A = dataset_alquileres.shape[0]:** Se asigna a la variable **A** el número total de filas en el DataFrame dataset_alquileres utilizando el atributo shape y seleccionando el elemento en la posición 0, que representa el número de filas.

**dropna(subset=["Valor"]):** Este método se utiliza para eliminar filas que contienen valores nulos. El argumento subset especifica en qué columnas se deben buscar los valores nulos. En este caso, se está buscando nulos solo en la columna **"Valor"**.

**inplace=True:** Este argumento indica que la modificación debe realizarse directamente en el df original (dataset_alquileres) en lugar de devolver un nuevo DataFrame modificado. Si inplace se establece en True, no es necesario asignar el resultado a una nueva variable; el df original se actualizará directamente.

**B = dataset_alquileres.shape[0]:** Se asigna a la variable **B** el nuevo número total de filas en el DataFrame después de aplicar la condición de filtro.

**A - B:** Se calcula la diferencia entre el número total de filas antes **(A)** y después **(B)** de aplicar el filtro. Esta diferencia representa el número de filas eliminadas por la condición.

En resumen, el código está calculando cuántas filas fueron eliminadas del DataFrame **dataset_alquileres** al aplicar la condición => dataset_alquileres.dropna(subset= ["Valor"], inplace= True)

```python
A = dataset_alquileres.shape[0]
dataset_alquileres.dropna(subset= ["Valor"], inplace= True)
B = dataset_alquileres.shape[0]
A - B
```

Ya no tengo filas con valores nulos en la columna valor.

```python
dataset_alquileres[dataset_alquileres["Valor"].isnull()]
```

**Tratamiento de Datos Faltantes II**

```python
dataset_alquileres
```

Filtro todos los registros que tienen valores nulos en la columna mantenimiento.

```python
dataset_alquileres[dataset_alquileres["Mantenimiento"].isnull()]
```

Todo tipo de registros que tienen valor nulo en la columna Mantenimiento-

```python
dataset_alquileres[dataset_alquileres["Mantenimiento"].isnull()].shape[0]
```

Filtrando por los Registros que presenten valor nulo(NaN) en la columna Mantenimiento y sean de tipo Departamento.

```python
sel_mantenimiento_deptos = (dataset_alquileres["Mantenimiento"].isnull()) & \
(dataset_alquileres["Tipo"]=="Departamento")
```

**dataset_alquileres = dataset_alquileres[~sel_mantenimiento_deptos]:** Se filtran las filas del DataFrame dataset_alquileres usando la condición **inversa** de sel_mantenimiento_deptos mediante el operador **~.** Esto significa que se está seleccionando las filas que no cumplen con la condición especificada en sel_mantenimiento_deptos.

```python
A = dataset_alquileres.shape[0]
dataset_alquileres = dataset_alquileres[~sel_mantenimiento_deptos]
B = dataset_alquileres.shape[0]
A - B
```

```python
dataset_alquileres[dataset_alquileres["Mantenimiento"].isnull()].shape[0]
```

**fillna({"Mantenimiento": 0, "Impuesto": 0}):** Este método se utiliza para llenar los valores nulos en el DataFrame. El argumento es un diccionario donde las claves son los nombres de las columnas y los valores son los valores con los que se quiere llenar los nulos. En este caso, estoy llenando los nulos en las columnas "Mantenimiento" e "Impuesto" con el valor 0 en ambos casos.

Después de ejecutar esta línea de código, las filas que originalmente tenían valores nulos en las columnas "Mantenimiento" e "Impuesto" ahora contendrán 0 en lugar de nulos. El DataFrame dataset_alquileres se modificará directamente.

```python
dataset_alquileres = dataset_alquileres.fillna({"Mantenimiento":0, "Impuesto":0})
```

```python
dataset_alquileres[dataset_alquileres["Mantenimiento"].isnull()].shape[0]
```

```python
dataset_alquileres[dataset_alquileres["Impuesto"].isnull()].shape[0]
```

El df dataset_alquileres ya no tiene valores nulos.
