Arreglando los valores nulos.

```python
df_residencial['Impuesto'] = df_residencial['Impuesto'].fillna(0)
df_residencial['Mantenimiento'] = df_residencial['Mantenimiento'].fillna(0)
```

#**Creando Nuevas Variables**

Creando una nueva columna llamada Valor Bruto para almacenar el valor de la suma de las columnas Valor, Mantenimiento e Impuesto para cada fila.

```python
df_residencial["Valor Bruto"] = (df_residencial["Valor"]) +\
(df_residencial["Mantenimiento"]) +\
(df_residencial["Impuesto"])
```

```python
df_residencial.head(10)
```

Obteniendo el valor por metro cuadrado.

```python
df_residencial["Valor m2"] = (df_residencial["Valor"]) / (df_residencial["Area"])
df_residencial.head(10)
```

Redondeo el valor de la columna Valor m2 a solo 2 decimales para una mejor visualizacion.

```python
df_residencial["Valor m2"] = df_residencial["Valor m2"].round(2)
df_residencial.head(10)
```

```python
df_residencial
```

  Obteniendo el valor bruto por metro cuadrado.

```python
df_residencial["Valor Bruto m2"] = ((df_residencial["Valor Bruto"]) / (df_residencial["Area"])).round(2)
df_residencial.head(10)
```

Creando una nueva columna que se encargara de agrupar todos los registros que sean algun tipo de casa.

```python
casa = ["Casa", "Casa en condominio", "Casa de villa"]
```

```python
# Defino una función que agrupa los tipos de casa
def agrupar_tipo(item):
    if "Casa" in item:
        return "CASA"
    else:
        return "Otros"
```

```python
# Aplicar la función a la columna "Tipo" y crear la nueva columna "Tipo de Casa"
df_residencial["Tipo de Casa"] = df_residencial["Tipo"].apply(agrupar_tipo)
```
