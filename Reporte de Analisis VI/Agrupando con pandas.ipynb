#**Creando Agrupamientos**

```python
barrios_agrupados = datos.groupby("Distrito")
```

Un tipo especifico de df (group by).

```python
type(barrios_agrupados)
```

El metodo groups me crea un diccionario donde las llaves son los nombres de los barrios y sus respectivos valores son las posiciones(indexes) donde aparecen.

```python
barrios_agrupados.groups
```

Con B me recorre los barrios y con D me genera un df para cada barrio.

```python
for B, D in barrios_agrupados:
  print(D)
```

Calculo el promedio de la columna Valor para cada uno de los barrios del df barrios_agrupados.

```python
for B1, D1 in barrios_agrupados:
  print("{} => {}".format(B1, D1.Valor.mean()))
```

Mismo resultado que con el ciclo for pero de una manera mas sencilla y redondeando los numeros a 2 decimales.

```python
barrios_agrupados["Valor"].mean().round(2)
```

Me genera un df con la media de las columnas que le indique para cada barrio.

```python
barrios_agrupados[["Valor", "Mantenimiento"]].mean().round(2)
```

#**Estadisticas Descriptivas con describe() y aggregate()**

Generando algunas estadisticas descriptivas a traves del metodo describe.

```python
barrios_agrupados["Valor"].describe().round(2)
```

El método **aggregate()** se utiliza para aplicar una o más funciones de agregación(sum(), mean(), count(), max(), min()) a los datos en un DataFrame. Se Puede pasar una sola función de agregación o una lista de funciones.

Generando mis propias estadisticas.

```python
barrios_agrupados["Valor"].aggregate(["min", "max"]).rename(columns={"min": "Minimo", "max":"Maximo"})
```

#**Explorando la libreria Matplotlib I**

Cuando se usa **%matplotlib** inline, los gráficos generados por Matplotlib se incrustan directamente en el cuaderno, lo que significa que se mostrarán en la celda de salida del cuaderno sin necesidad de abrir ventanas emergentes o ventanas separadas para ver los gráficos.

Sin esta línea de código, los gráficos de Matplotlib se mostrarían en una ventana emergente por defecto, lo que podría no ser tan conveniente en un entorno como Jupyter.

En resumen, **%matplotlib inline** habilita la representación en línea de gráficos generados por Matplotlib en un cuaderno Jupyter, lo que facilita la visualización y el análisis de datos en el mismo documento.

**rc:** Recurso.

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc("figure", figsize = (20,10))  #creando una figura grafica y dandole dimensiones
```

Promedio del valor de los inmuebles por distrito.

```python
grafico = barrios_agrupados["Valor"].mean().plot.bar(color = "Magenta")
grafico.set_ylabel("Valor del Alquiler")
grafico.set_title("Valor Medio del Alquiler por Distrito", {"fontsize" : 20})
```

Valor Maximo de inmuebles por distrito.

**Outliers:** Hay algunos datos que no tienen sentido en el grafico(en el anterior grafico tambien pasa lo mismo) como el valor maximo del distrito Lince en este grafico de valores maximos de alquileres. Estos valores irracionales o fuera de foco se llaman Outliers.

```python
grafico = barrios_agrupados["Valor"].max().plot.bar(color = "Magenta")
grafico.set_ylabel("Valor del Alquiler")
grafico.set_title("Valor Maximo del Alquiler por Distrito", {"fontsize" : 20})
```
